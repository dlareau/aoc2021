```python
import aocd
puzzle = aocd.models.Puzzle(year=2021, day=2)
data = puzzle.input_data.split("\n")
data = [x.split(" ") for x in data]
```

```python
data
```

```python
depth = 0
pos = 0
for d in data:
    if(d[0] == "down"):
        depth += int(d[1])
    elif(d[0] == "up"):
        depth -= int(d[1])
    elif(d[0] == "forward"):
        pos += int(d[1])
puzzle.answer_a = pos*depth
```

```python
depth = 0
pos = 0
aim = 0
for d in data:
    if(d[0] == "down"):
        aim += int(d[1])
    elif(d[0] == "up"):
        aim -= int(d[1])
    elif(d[0] == "forward"):
        pos += int(d[1])
        depth += aim * int(d[1])
puzzle.answer_b = pos*depth
```
